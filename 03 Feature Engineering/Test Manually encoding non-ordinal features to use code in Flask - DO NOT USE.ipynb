```python
if (bankchurners20columns.loc[bankchurners20columns.index[0], 'Gender'] == "M"):
    bankchurners20columns.loc[bankchurners20columns.index[0], 'Gender'] = 1
```

```python
#Gender
for i in range(len(bankchurners20columns)):
    if (bankchurners20columns.loc[bankchurners20columns.index[i], 'Gender'] == "F"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Gender'] = 0
    elif (bankchurners20columns.loc[bankchurners20columns.index[i], 'Gender'] == "M"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Gender'] = 1
        
bankchurners20columns['Gender'] = bankchurners20columns['Gender'].astype(int)   
```

```python
#Marital_Status
for i in range(len(bankchurners20columns)):
    if (bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] == "Married"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] = 1
    elif (bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] == "Single"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] = 2
    elif (bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] == "Unknown"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] = 3
    elif (bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] == "Divorced"):
        bankchurners20columns.loc[bankchurners20columns.index[i], 'Marital_Status'] = 4
        
bankchurners20columns['Marital_Status'] = bankchurners20columns['Marital_Status'].astype(int) 
```

```python
testdata = pd.read_csv('../data_source/TestBulkData.csv')
testdata.head(10)
```

```python
#Gender
for i in range(len(testdata)):
    if (testdata.loc[testdata.index[i], 'Gender'] == "F"):
        testdata.loc[testdata.index[i], 'Gender'] = 0
    elif (testdata.loc[testdata.index[i], 'Gender'] == "M"):
        testdata.loc[testdata.index[i], 'Gender'] = 1
        
testdata['Gender'] = testdata['Gender'].astype(int) 
```

```python
#Marital_Status
for i in range(len(testdata)):
    if (testdata.loc[testdata.index[i], 'Marital_Status'] == "Married"):
        testdata.loc[testdata.index[i], 'Marital_Status'] = 1
    elif (testdata.loc[testdata.index[i], 'Marital_Status'] == "Single"):
        testdata.loc[testdata.index[i], 'Marital_Status'] = 2
    elif (testdata.loc[testdata.index[i], 'Marital_Status'] == "Unknown"):
        testdata.loc[testdata.index[i], 'Marital_Status'] = 3
    elif (testdata.loc[testdata.index[i], 'Marital_Status'] == "Divorced"):
        testdata.loc[testdata.index[i], 'Marital_Status'] = 4
        
testdata['Marital_Status'] = testdata['Marital_Status'].astype(int) 
```

```python
map_education = {"Unknown":1, "Uneducated":2, "High School":3, "College":4, "Graduate":5, "Post-Graduate":6, "Doctorate":7}
testdata['Education_Level'] = testdata['Education_Level'].map(map_education)

map_income = {"Unknown":1, "Less than $40K" : 2, "$40K - $60K": 3, "$60K - $80K": 4, "$80K - $120K": 5, "$120K +":6}
testdata['Income_Category'] = testdata['Income_Category'].map(map_income)

map_cards = {"Blue":1, "Silver" : 2, "Gold": 3, "Platinum": 4}
testdata['Card_Category'] = testdata['Card_Category'].map(map_cards)
```
